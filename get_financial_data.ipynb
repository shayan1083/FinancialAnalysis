```python
import numpy as np
import pandas as pd
import yfinance as yf
import mysql.connector as sql
from sqlalchemy import create_engine
```

```python
msft = yf.Ticker("MSFT")
```

```python
msft.info
```

```python
hist = msft.history(period="10y")
```

```python
hist
```

```python
type(hist)
```

```python
fin_data = pd.DataFrame(hist)
fin_data.drop(columns=['Dividends','Stock Splits'], inplace=True)
```

```python
fin_data
```

```python
fin_data.index
```

```python
fin_data.index = pd.to_datetime(fin_data.index).date
```

```python
fin_data.index
```

```python
fin_data
```

```python
engine = create_engine("mysql+mysqlconnector://root:password@localhost/stock_data")
```

```python
df = fin_data.reset_index()
df.columns = ['record_date','open_price','high_price','low_price','close_price','volume']
```

```python
df
```

```python
df.to_sql(name='financial_data', con=engine, if_exists='append', index=False)
```
